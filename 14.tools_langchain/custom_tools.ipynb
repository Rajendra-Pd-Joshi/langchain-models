```python
# step 1 define a function
def multiply(a,b):
    return a*b
```

```python
# step 2 add description and provide typedecorators
def multiply(a:int,b:int)->int:
    'this is a multiply function that multiply two integers'
    return a*b
```

```python
# step 3 always use the @tool decorator
@tool
def multiply(a:int,b:int)->int:
    'this is a multiply function that multiply two integers'
    return a*b
```

```python
multiply.invoke({'a':2,'b':3})
```

## Method-2 :using StructuredTool & Pydantic

```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class MultiplyInput(BaseModel):
    a:int = Field(required=True,description='first number')
    b:int = Field(required=True,description='second number')
```

```python
def multiply_func(a:int,b:int)->int:
    return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func = multiply_func,
    description='this function multiply two integers',
    args_schema = MultiplyInput,
    name='multiply'
)
```

```python
multiply_tool.invoke({'a':10,'b':23})
```

```python
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

## Method 3: BaseTool

```python
from langchain.tools import BaseTool
from typing import Type
from pydantic import BaseModel,Field
```

```python
# arg Schema using pydantic
class MultiplyInput(BaseModel):
    a:int = Field(required=True,description='first number')
    b:int = Field(required=True,description='second number')
```

```python
class MultiplyTool(BaseTool):
    name:str = 'multiply'
    description:str = 'multiply two integers'
    args_schema:Type[BaseModel]=MultiplyInput

    def _run(self,a:int,b:int)->int:
        return a*b
```

```python
multiply_tool = MultiplyTool()
```

```python
multiply_tool.invoke({'a':3,'b':6})
```

```python
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```
