```python
from langchain_text_splitters import RecursiveCharacterTextSplitter,Language
```

```python
splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.PYTHON,
    chunk_size=400,
    chunk_overlap = 40
)
```

```python
code = """def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False
    return True
    if __name__ == "__main__":
    # Check a few numbers
    for num in [7, 10, 17, 100, 97]:
        print(f"{num} is prime: {is_prime(num)}")

    print()

    # Primes up to 50
    print("Primes up to 50:", sieve_of_eratosthenes(50))

    print()

    # First 10 primes
    print("First 10 primes:", first_n_primes(10))"""
```

```python
docs = splitter.split_text(code)
```
