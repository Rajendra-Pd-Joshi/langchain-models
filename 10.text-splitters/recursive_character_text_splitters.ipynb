## Recusrsive Character Text Splitters

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
sample_text = """LangChain is a framework for developing applications powered by large language models (LLMs). It simplifies every stage of the LLM application lifecycle, from development to deployment.

For development, LangChain provides building blocks and components for building applications using its open-source building blocks and components. You can use LangChain's components to build new chains, or use existing chains out of the box.

Document loading and splitting is a critical step in any RAG pipeline. Before you can embed and retrieve relevant chunks, you need to break large documents into smaller, semantically meaningful pieces.
The RecursiveCharacterTextSplitter is the recommended way to do this for generic text.
It tries to split on paragraph breaks first, then newlines, then spaces, and finally individual characters if nothing else works.

Chunk overlap is another important parameter. It ensures that some context from the end of one chunk carries over into the beginning of the next chunk, which helps preserve continuity when a sentence or idea spans a chunk boundary.

Different document types call for different splitting strategies. Markdown files benefit from header-aware splitters. Code files benefit from language-aware splitters that respect function and class boundaries. Plain prose generally works fine with the default recursive splitter."""
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 20,
    separators=['\n\n','\n',' ',''] # by default it is like this
)
```

```python
docs = splitter.split_text(sample_text)
```

## Recursive Character Text Splitter for the docuement

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
```

```python
Loader = PyPDFLoader("dl-curriculum (1).pdf")
```

```python
document = Loader.load()
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 20,
    separators=['\n\n','\n',' ','']
)
```

```python
splitted_docs = splitter.split_documents(document)
```
