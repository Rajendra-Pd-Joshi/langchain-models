```python
from langchain_experimental.text_splitter import SemanticChunker
```

```python
from langchain_openai import OpenAIEmbeddings
```

```python
embedding_object = OpenAIEmbeddings()
```

```python
splitter_object = SemanticChunker(embeddings=embedding_object,breakpoint_threshold_type='standard_deviation',breakpoint_threshold_amount=1)
```

```python
text = """Cats are wonderful pets.

They enjoy sleeping.

They like fish.

Quantum computing uses qubits.

Entanglement is important.

Quantum computers solve different problems."""
```

```python
documents = splitter_object.create_documents([text])
```
