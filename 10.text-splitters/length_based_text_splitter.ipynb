## CharacterTextSplitter with raw text

```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
splitter = CharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 20,
    separator=''
)
```

```python
text = ''' Claude token limits vary by access method, with subscription plans capped by a 5-hour rolling window and API users limited by per-minute rates. 

Subscription Plans (Claude Pro/Max)
Usage is shared across claude.ai, Claude Code, and Claude Desktop, measured in cost-based tokens rather than raw counts:

Pro ($20/mo): ~44,000 tokens per 5-hour window. 
Max 5x ($100/mo): ~88,000 tokens per 5-hour window. 
Max 20x ($200/mo): ~220,000 tokens per 5-hour window. 
Weekly Caps: Additional weekly limits apply across all models, with separate caps for Sonnet usage. 
Context Window: 200K tokens for all plans (except Enterprise, which has 500K). '''
```

```python
docs = splitter.split_text(text)
```

## Now integrating the document loader and the characterTextSplitters


```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter
```

```python
Loader = PyPDFLoader(file_path='dl-curriculum (1).pdf')
```

```python
docs = Loader.load()
```

```python
splitter1 = CharacterTextSplitter(
    chunk_size = 100,
    chunk_overlap = 20,
    separator=''
)
```

```python
docs_splitted = splitter1.split_documents(docs)
```

```python
docs_splitted[0]
```
