### MMR (Maximal Marginal Relevance) is a reranking algorithm used in LangChain retrievers to optimize for both similarity to the query and diversity among the results


```python
import os
from dotenv import load_dotenv
from langchain_openai.embeddings import OpenAIEmbeddings
```

```python
embeddings = OpenAIEmbeddings(
    model = os.getenv('EMBEDDING_MODEL'),
    api_key= os.getenv('OPENAI_API_KEY')
)
```

### Use the FAISS retriver model

```python
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document
```

```python
# Sample documents
docs = [
    Document(page_content="LangChain makes it easy to work with LLMs."),
    Document(page_content="LangChain is used to build LLM based applications."),
    Document(page_content="Chroma is used to store and search document embeddings."),
    Document(page_content="Embeddings are vector representations of text."),
    Document(page_content="MMR helps you get diverse results when doing similarity search."),
    Document(page_content="LangChain supports Chroma, FAISS, Pinecone, and more."),
]
```

```python
vector_store = FAISS.from_documents(
    documents=docs,
    embedding=embeddings
)
```

```python
# Enable MMR as retriver 
retriver = vector_store.as_retriever(
    search_type = 'mmr',
    search_kwargs ={'k':3,'lambda_mult':1} # just same as that of similariy search because lambda_mult =1
)
```

```python
type(retriver)
```

```python
result = retriver.invoke('what is langchain?')
```

```python
for i,doc in enumerate(result):
    print(f'document {i+1}')
    print(doc.page_content)
```

```python
retriver1 = vector_store.as_retriever(
    search_type = 'mmr',
    search_kwargs = {'k':3,'lambda_mult':0.5}
)
```

```python
result1 = retriver1.invoke('what is langchain?')
```

```python
for i,doc in enumerate(result1):
    print(f'document {i+1}')
    print(doc.page_content)
```
