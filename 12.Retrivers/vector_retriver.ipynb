## Vector Retrivers in langchain

```python
import os
from dotenv import load_dotenv
from langchain_openai import OpenAI
from langchain_openai.embeddings import OpenAIEmbeddings
```

```python
load_dotenv()
```

### Initialize the embedding model

```python
embeddings = OpenAIEmbeddings(
    model = os.getenv('EMBEDDING_MODEL'),
    base_url= os.getenv('OPENAI_BASE_URL'),
    api_key=os.getenv('OPENAI_API_KEY')
)
```

### Step 1: Your source documents

```python
from langchain_core.documents import Document
```

```python
documents = [
    Document(page_content="LangChain helps developers build LLM applications easily."),
    Document(page_content="Chroma is a vector database optimized for LLM-based search."),
    Document(page_content="Embeddings convert text into high-dimensional vectors."),
    Document(page_content="OpenAI provides powerful embedding models."),
]
```

### Create Chroma vector store in Memory

```python
from langchain_chroma import Chroma
```

```python
vector_stores = Chroma.from_documents(
    documents = documents,
    embedding = embeddings,
    collection_name = 'my_collection'
)
```

```python
vector_stores.get()
```

```python
# Converting the vector store into a retrivers
retriver = vector_stores.as_retriever(search_kwargs={'k': 2})
```

```python
query = "What is Chroma used for?"
```

```python
results = retriver.invoke(query)
```

```python
for i,doc in enumerate(results):
    print(f'document {i+1}')
    print(doc.page_content)
```

### Similar type of output we can get from the vector_store.similarity_search()

```python
results = vector_stores.similarity_search(query='what is chroma used for?',k=2)
```

```python
for i,doc in enumerate(results):
    print(f'document {i+1}')
    print(doc.page_content)
```
